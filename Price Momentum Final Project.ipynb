```python
# Price momentum calculators
import pandas as pd
import numpy as np

class Tools:
    def __init__(self, ivv_df, Tickers):
        self.price=ivv_df["Adj Close"]
        self.vol=ivv_df["Volume"]
        self.tckrs=Tickers
        
        
        self.mtbl= pd.DataFrame(index = self.tckrs)
      

    def calculate_percent_above_260_low(self):
        percent_above_260_low = (self.price - self.price.rolling(window=260, min_periods=1).min()) / self.price.rolling(window=260, min_periods=1).min()
        self.mtbl["percent_above_260_low"] = percent_above_260_low.mean()

    def calculate_4_52_week_price_oscillator(self):
        short_ma = self.price.rolling(window=4 * 5, min_periods=1).mean()  # 4 weeks (5 trading days per week)
        long_ma = self.price.rolling(window=52 * 5, min_periods=1).mean()  # 52 weeks
        self.mtbl["4_52_week_price_oscillator"] = ((short_ma - long_ma) / long_ma).mean()

    def calculate_52_week_trend_slope(self):
       self.mtbl["52 Week Trend"] = self.price.rolling(252).sum()[-20:].mean()

    def calculate_39_week_return(self):
        return_39_week = (self.price / self.price.shift(39 * 5) - 1).mean()  
        self.mtbl["39_week_return"] = return_39_week

    def calculate_51_week_volume_price_trend(self):
        self.mtbl["Price Volume"] = ((self.price.pct_change() * self.vol).rolling(252).sum()[-20:].mean())

    def get_momentum_factors(self):
        self.calculate_percent_above_260_low()
        self.calculate_52_week_trend_slope()
        self.calculate_4_52_week_price_oscillator()
        self.calculate_39_week_return()
        self.calculate_51_week_volume_price_trend()

        # Return the mean of the calculated factors along rows
        return self.mtbl
    
    def normalizing(self):
        # Z-score normalization
        normalized_factors = (self.get_momentum_factors() - self.mtbl.mean()) / self.mtbl.std()

        # Aggregate the scores (sum across rows)
        z_scores = normalized_factors.sum(axis=1)
        return z_scores
    
    def baskets(self):
        self.z_scored = self.normalizing()
        return self.z_scored.nlargest(10).index, self.z_scored.nsmallest(10).index
       
    
tools_instance = Tools(ivv_df=ivv_data, Tickers=ticker)
momentum_factors = tools_instance.get_momentum_factors()
zscore = tools_instance.normalizing()
baskets_result = tools_instance.baskets()  # Rename the variable

display(momentum_factors)
display(zscore)
display(baskets_result)
```

```python
import yfinance as yf
import pandas as pd
import datetime as dt

data = yf.download(ticker, period='10y')
data = data.dropna(how="all", axis=1)

hist_rets = data["Adj Close"].pct_change().resample("M").sum().shift(-1)

date_list = pd.DataFrame(pd.date_range("2017-06-01", dt.datetime.today(), freq="M"))
date_list.rename(columns={0: "date"}, inplace=True)

long_ret, shrt_ret = pd.DataFrame(), pd.DataFrame()


for date in date_list.date:
    MF = Tools(data[data.index <= date], ticker)
    longs, short = MF.baskets()

    long_ret = long_ret.append(hist_rets[longs].loc[date])
    shrt_ret = shrt_ret.append(hist_rets[short].loc[date])


```

```python
display(long_ret.mean(axis = 1))
display(shrt_ret.mean(axis = 1))
```

```python
tot_ret = pd.DataFrame(long_ret.mean(axis=1) - shrt_ret.mean(axis=1)).shift()
tot_ret = tot_ret[-60:]

tot_ret['Return[Cumulative]'] = np.exp(np.log1p(tot_ret).cumsum())

tot_ret['Bench'] = yf.download('IVV', period = '6y')['Adj Close'].pct_change().resample('M').sum()

tot_ret['IVV [Cumulative]'] = np.exp(np.log1p(tot_ret['Bench']).cumsum())

tot_ret['Is it Positive?'] = tot_ret[0] > 0

tot_ret['cVs'] = tot_ret['Return[Cumulative]'] - tot_ret['IVV [Cumulative]']
tot_ret['is it [Cumul.] Positive?'] = tot_ret['cVs'] > 0

tot_ret.index = tot_ret.index.date
tot_ret=tot_ret.round(3)
tot_ret
```

```python
pt1, a1 = plt.subplots(figsize = (20,5))
a1 = tot_ret[0].plot(kind = 'bar', color = tot_ret['Is it Positive?'].map({True: 'g', False: 'r'}))
plt.title('Monthly portfolio return bar chart IVV')
plt.show()
```

```python
pt2, a2 = plt.subplots(figsize = (20, 5))
af = a2.twinx()
X = tot_ret.index
af.fill_between(X, tot_ret['Return[Cumulative]'], tot_ret['IVV [Cumulative]'], alpha = 0.30)
a2.plot(X, tot_ret['Return[Cumulative]'], color= 'green')
a2.plot(X, tot_ret['IVV [Cumulative]'], color= 'red')
plt.show()
```
